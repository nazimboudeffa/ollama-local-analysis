# 📊 Trading Signal Analysis with Ollama + Qwen
Test local LLM for trading signals

```python
import requests
import json

OLLAMA_URL = "http://localhost:11434/api/generate"
MODEL = "qwen3:4b"
```

```python
def analyze_signal(data):
    prompt = f"""
You are a professional trading analyst.

Return STRICT JSON:
{{
  "signal": "BUY | SELL | HOLD",
  "confidence": 0-100,
  "reason": "short"
}}

Data:
{json.dumps(data)}
"""

    response = requests.post(
        OLLAMA_URL,
        json={
            "model": MODEL,
            "prompt": prompt,
            "stream": False,
            "options": {
                "temperature": 0.2,
                "num_predict": 150
            }
        },
        timeout=30
    )

    return response.json()
```

```python
# Example trading data
data = {
    "pair": "BTC/USDT",
    "timeframe": "5m",
    "price": 67250,
    "rsi": 78,
    "macd": "bearish crossover",
    "volume": "high",
    "trend": "uptrend"
}

result = analyze_signal(data)
print(result)
```

## 🔁 Multi-timeframe example

```python
data_multi = {
    "BTC/USDT": {
        "1m": {"rsi": 80, "trend": "up"},
        "5m": {"rsi": 72, "trend": "up"},
        "15m": {"rsi": 65, "trend": "neutral"}
    }
}

print(analyze_signal(data_multi))
```
